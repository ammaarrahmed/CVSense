Importing libraries

```python
import os
from sklearn.feature_extraction.text import TfidfVectorizer
import pickle
```

Loading Module 2 output files

```python
from google.colab import files
uploaded = files.upload()
```

```python
import pandas as pd
```

```python
import io
preprocessed_resumes = pd.read_csv(io.BytesIO(uploaded['preprocessed_resumes_final.csv']))
preprocessed_jobs =pd.read_csv(io.BytesIO(uploaded['preprocessed_jobs_final (1).csv']))
```

```python
preprocessed_resumes.head()
```

```python
preprocessed_jobs.head()
```

Extracting Preprocessed Text Column

```python
resume_texts = preprocessed_resumes["preprocessed_text"].astype(str).tolist()
jd_texts = preprocessed_jobs["preprocessed_text"].astype(str).tolist()
```

Combining Corpus

```python
all_texts = resume_texts + jd_texts
```

Implementing TF-IDF Vectorizer

```python
vectorizer = TfidfVectorizer(
    max_features=5000,
    ngram_range=(1, 2),
    stop_words="english"
)

tfidf_matrix = vectorizer.fit_transform(all_texts)
```

Splitting Vectors

```python
resume_vectors = tfidf_matrix[:len(resume_texts)]
jd_vectors = tfidf_matrix[len(resume_texts):]
```

Saving Output

```python
import os
import pickle

os.makedirs("outputs", exist_ok=True)

output = {
    "resume_vectors": resume_vectors,
    "jd_vectors": jd_vectors,
    "feature_names": vectorizer.get_feature_names_out()
}

with open("outputs/tfidf_vectors.pkl", "wb") as f:
    pickle.dump(output, f)

```

Looking Vectors Output
